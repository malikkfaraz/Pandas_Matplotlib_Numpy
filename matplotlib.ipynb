```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
days = list(range(0,22,3))
t_min = [20,25,27,21,24,30,25,24]
t_max = [26,29,30,29,29,36,31,30]
plt.plot(days,t_min ,'b')
plt.plot(days,t_min ,'ob')
plt.plot(days,t_max ,'r')
plt.plot(days,t_max ,'or')
plt.xlabel('days---->')
plt.ylabel('temperature ----->')
plt.show()
```
